```python
import pandas as pd
import os
```

```python
import re
from collections import defaultdict

filepath = '../Nov.25__smartlog.scrapper/smlog_data'

# 1. 디렉토리 내의 모든 csv 파일 찾기
csv_files = [f for f in os.listdir(filepath) if f.endswith('.csv')]

# 2. 파일명에서 날짜(YYYY-MM-DD)와 변수 추출 (예: '2025-08-18_네트워크.csv' 식)
# 날짜는 무시하고 변수명만으로 그룹화
pattern = re.compile(r"\d{4}-\d{2}-\d{2}_(.+)\.csv")

grouped_files = defaultdict(list)
for fname in csv_files:
    m = pattern.match(fname)
    if m:
        var = m.group(1)  # 변수명만 추출 (날짜는 무시)
        grouped_files[var].append(os.path.join(filepath, fname))

# 3. 각 변수별로 합치기 (날짜 무시)
for var_name, files in grouped_files.items():
    dfs = [pd.read_csv(f) for f in files]
    merged = pd.concat(dfs, ignore_index=True)
    output_path = os.path.join(filepath, f"{var_name}_merged.csv")
    merged.to_csv(output_path, index=False, encoding="utf-8-sig")
    print(f"Saved merged file: {output_path} ({len(files)} files merged)")
```
